```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(34) #random rakamlar buna göre üretilecek
x = np.arange(20)
y = np.random.normal(10,1,20) 
#ortalaması 10, standart sapması 1 olmak üzere 20 adet rakam üret
z = np.random.normal(10,1,20) 
k = np.random.normal(10,3,20)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,k)
```

```python
fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(311)
ax.set_title("x'in y'ye göre değişimi")
ax.plot(x,y)

bx = fig.add_subplot(312)
bx.set_title("x'in z'ye göre değişimi")
bx.plot(x,z)

cx = bx = fig.add_subplot(313)
cx.set_title("x'in k'ye göre değişimi")
cx.plot(x,k)
```

#Feature engineering - verinin aslında olmayan ancak veride yararlanılıp üretilen sütun işlemi
#Öznitelik Mühendsiliği

```python
df["Date"] = pd.to_datetime(df['Date'])
```

```python
df['day'] = (df['Date']).dt.day
df['month'] = (df['Date']).dt.month
df['year'] = (df['Date']).dt.year 
#sonradan oluşturduklarımız feature engineering
```

```python
df.info()
```

```python
df['month_name'] = df['Date'].dt.month_name()
df['weekday'] = df['Date'].dt.day_name()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['Hour'] = (df['Time']).dt.hour
```
